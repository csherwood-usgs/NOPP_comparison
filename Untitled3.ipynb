```python
import io
import gzip
import requests
import numpy as np
import pandas as pd
from pathlib import Path

# ============================================================
# USER SETTINGS
# ============================================================
station = "42099"
start = "2024-09-24"
end   = "2024-09-29"
year  = 2024

out_dir = Path("F:/crs/proj/2025_NOPP_comparison/helene_waves")
out_dir.mkdir(parents=True, exist_ok=True)

hist_url = f"https://www.ndbc.noaa.gov/data/historical/stdmet/{station}h{year}.txt.gz"
rt_url   = f"https://www.ndbc.noaa.gov/data/realtime2/{station}.txt"

# ============================================================
# HELPERS
# ============================================================
def parse_ndbc_table_text(text, station):
    lines = [ln.rstrip() for ln in text.splitlines() if ln.strip()]
    if len(lines) < 2:
        raise ValueError(f"{station}: file is empty or malformed")

    header_line = lines[0].lstrip("#").strip()
    cols = header_line.split()

    data_start = 1
    if len(lines) > 1:
        second_tokens = lines[1].split()
        if second_tokens and not second_tokens[0].replace("-", "").isdigit():
            data_start = 2

    data_text = "\n".join(lines[data_start:])
    if not data_text.strip():
        raise ValueError(f"{station}: no data rows found")

    df = pd.read_csv(
        io.StringIO(data_text),
        sep=r"\s+",
        names=cols,
        na_values=["MM", "999", "999.0", "9999", "9999.0", "99.00", "999.00"],
        engine="python",
    )

    # Historical files may use YY or YYYY
    if {"YY", "MM", "DD", "hh", "mm"}.issubset(df.columns):
        yy = pd.to_numeric(df["YY"], errors="coerce")
        year_vals = np.where(yy < 100, yy + 1900, yy)
        df["time"] = pd.to_datetime(
            {
                "year": year_vals,
                "month": pd.to_numeric(df["MM"], errors="coerce"),
                "day": pd.to_numeric(df["DD"], errors="coerce"),
                "hour": pd.to_numeric(df["hh"], errors="coerce"),
                "minute": pd.to_numeric(df["mm"], errors="coerce"),
            },
            errors="coerce",
            utc=True,
        )
    elif {"YYYY", "MM", "DD", "hh", "mm"}.issubset(df.columns):
        df["time"] = pd.to_datetime(
            {
                "year": pd.to_numeric(df["YYYY"], errors="coerce"),
                "month": pd.to_numeric(df["MM"], errors="coerce"),
                "day": pd.to_numeric(df["DD"], errors="coerce"),
                "hour": pd.to_numeric(df["hh"], errors="coerce"),
                "minute": pd.to_numeric(df["mm"], errors="coerce"),
            },
            errors="coerce",
            utc=True,
        )
    else:
        raise ValueError(f"{station}: could not identify time columns")

    for c in ["WDIR", "WSPD", "GST", "WVHT", "DPD", "APD", "MWD",
              "PRES", "ATMP", "WTMP", "DEWP", "VIS", "PTDY", "TIDE"]:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    df["station"] = station
    return df


def subset_date_range_inclusive(df, start, end):
    df = df.copy()
    df["time"] = pd.to_datetime(df["time"], utc=True)

    t0 = pd.Timestamp(start, tz="UTC")
    t1 = pd.Timestamp(end, tz="UTC") + pd.Timedelta(days=1) - pd.Timedelta(seconds=1)

    return (
        df[(df["time"] >= t0) & (df["time"] <= t1)]
        .sort_values("time")
        .reset_index(drop=True)
    )


def download_hist_stdmet_station_year(station, year):
    url = f"https://www.ndbc.noaa.gov/data/historical/stdmet/{station}h{year}.txt.gz"
    r = requests.get(url, timeout=120)

    if r.status_code == 404:
        raise FileNotFoundError(f"{station}: no historical stdmet file for {year}")

    r.raise_for_status()

    with gzip.GzipFile(fileobj=io.BytesIO(r.content)) as gz:
        text = gz.read().decode("utf-8", errors="replace")

    return parse_ndbc_table_text(text, station)


def download_realtime_station(station):
    url = f"https://www.ndbc.noaa.gov/data/realtime2/{station}.txt"
    r = requests.get(url, timeout=120)

    if r.status_code == 404:
        raise FileNotFoundError(f"{station}: no realtime2 file")

    r.raise_for_status()
    return parse_ndbc_table_text(r.text, station)


# ============================================================
# RUN
# ============================================================
print("Checking historical file:")
print(hist_url)

try:
    df = download_hist_stdmet_station_year(station, year=year)
    source_used = "historical"
except Exception as e:
    print(f"Historical download failed: {e}")
    print("\nTrying realtime fallback:")
    print(rt_url)
    df = download_realtime_station(station)
    source_used = "realtime2"

df = subset_date_range_inclusive(df, start, end)

print(f"\nSource used: {source_used}")
print(f"Records in window: {len(df)}")

if len(df) == 0:
    print("No records in requested time window.")
else:
    print("Columns:", df.columns.tolist())

    summary = {
        "station": station,
        "n_records": len(df),
        "time_min": df["time"].min(),
        "time_max": df["time"].max(),
        "has_WVHT": ("WVHT" in df.columns and df["WVHT"].notna().any()),
        "has_DPD":  ("DPD"  in df.columns and df["DPD"].notna().any()),
        "has_APD":  ("APD"  in df.columns and df["APD"].notna().any()),
        "has_MWD":  ("MWD"  in df.columns and df["MWD"].notna().any()),
        "max_WVHT_m": df["WVHT"].max(skipna=True) if "WVHT" in df.columns else np.nan,
    }

    print("\nSummary:")
    for k, v in summary.items():
        print(f"{k}: {v}")

    outfile = out_dir / f"ndbc_{station}_check.csv"
    df.to_csv(outfile, index=False)
    print(f"\nSaved: {outfile}")

    print("\nFirst few rows:")
    print(df.head())
```
